```python
import cv2
import sys
from src.utils import plot_random
import numpy as np
import glob
```

```python
no_tumor = []
no_path = r'../data/no/*.jpg'

for file in glob.iglob(no_path):
    img = cv2.imread(file)
    img = cv2.resize(img, (128, 128))
    b, g, r = cv2.split(img)
    img = cv2.merge([r, g, b])
    no_tumor.append(img)

tumor = []
yes_path = r'../data/yes/*.jpg'

for file in glob.iglob(yes_path):
    img = cv2.imread(file)
    img = cv2.resize(img, (128, 128))
    b, g, r = cv2.split(img)
    img = cv2.merge([r, g, b])
    tumor.append(img)

for i, img in enumerate(tumor):
    if i<5:
        print(img.shape)
    else: break
```

**See all the standardized data**

```python
healthy = np.array(no_tumor)
tumor = np.array(tumor)
```
